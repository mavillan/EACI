```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import numba
from math import sqrt
```

## Metric Definition

```python
#standar Euclidean distance
@numba.jit('float64 (float64[:], float64[:])', nopython=True)
def metric(x, y):
    n = len(x)
    ret = 0.
    for i in range(n):
        ret += (x[i]-y[i])**2
    return sqrt(ret)

@numba.jit('float64 (float64[:], float64[:,:])', nopython=True)
def d(x, Y):
    m = Y.shape[0]
    inf = np.inf
    
    for i in range(m):
        dist = metric(x, Y[i])
        if dist < inf:
            inf = dist
    return inf

# Hausdorff metric between two arrays
@numba.jit('float64 (float64[:,:], float64[:,:])', nopython=True)
def h(X, Y):
    m = X.shape[0]
    n = Y.shape[0]
    sup1 = -1.
    sup2 = -1.
    
    for i in range(m):
        inf1 = d(X[i], Y)
        if inf1 > sup1:
            sup1 = inf1
    for i in range(n):
        inf2 = d(Y[i], X)
        if inf2 > sup2:
            sup2 = inf2
            
    return max(sup1, sup2)
```

```python
"""
Automatically tune threshold parameters of Canny detector
"""
def auto_canny(image, sigma=0.33):
    # compute the median of the single channel pixel intensities
    v = np.median(image)
 
    # apply automatic Canny edge detection using the computed median
    lower = int(max(0, (1.0 - sigma) * v))
    upper = int(min(255, (1.0 + sigma) * v))
    edged = cv.Canny(image, lower, upper)
 
    # return the edged image
    return edged

"""
It blur it to supress noise, applys Canny edge detector,
and generate an 2D array in which each row has the position
where there exists a 1 bit.
"""
def image_to_array(image):
    #blur the image and compute edges
    image = cv.GaussianBlur(image, (3, 3), 0)
    edges = auto_canny(image)
    
    #generating the array with positions
    arr = np.where(edges == 255)
    arr = np.vstack(arr).T     
    return arr.astype(float)
```

```python
model_arr = image_to_array(model)
image_arr = image_to_array(image)
```

## Finding the pattern

```python
def find_pattern(model, image, slide = 5):
    #window size
    m,n = model.shape
    #image size
    M,N = image.shape
    #computing array representation of model
    model_arr = image_to_array(model)
    #minimal distance
    min_dist = np.inf
    #indexes of region with minimal distance (upper left corner)
    min_i = -1
    min_j = -1
    
    for i in range(0,M,slide):
        #don't go outside
        if i+m >= M: continue
        for j in range(0,N,slide):
            if j+n >= N: continue
            #window slice
            sub_image = image[i:i+m,j:j+n]
            #comuting array representation
            sub_image_arr = image_to_array(sub_image)
            #computing Hausdorff distance
            dist = h(model_arr, sub_image_arr)
            if dist < min_dist:
                min_dist = dist
                min_i = i
                min_j = j
    return min_dist, min_i, min_j
```

```python
dist, i, j = find_pattern(model, image)
print dist, i, j
```
